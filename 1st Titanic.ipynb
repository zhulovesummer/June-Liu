```python
import pandas as pd #数据分析
import numpy as np #科学计算
from pandas import Series,DataFrame

import os
os.chdir('D:/')
data_train = pd.read_csv('train.csv')
print(data_train.columns)

data_train.head()
```

```python
#特征数据预处理
'''通常遇到缺值的情况，我们会有几种常见的处理方式：

如果缺值的样本占总数比例极高，我们可能就直接舍弃了，作为特征加入的话，可能反倒带入noise，影响最后的结果了
如果缺值的样本适中，而该属性非连续值特征属性(比如说类目属性)，那就把NaN作为一个新类别，加到类别特征中
如果缺值的样本适中，而该属性为连续值特征属性，有时候我们会考虑给定一个step(比如这里的age，我们可以考虑每隔2/3岁为一个步长)，然后把它离散化，之后把NaN作为一个type加到属性类目中。
有些情况下，缺失的值个数并不是特别多，那我们也可以试着根据已有的值，拟合一下数据，补充上。
本例中，后两种处理方式应该都是可行的，我们先试试拟合补全吧(虽然说没有特别多的背景可供我们拟合，这不一定是一个多么好的选择)
我们这里用scikit-learn中的RandomForest来拟合一下缺失的年龄数据'''
```

```python
#Cabin数据根据有无做二值化处理
data_train["Cabin"]=data_train["Cabin"].fillna(0)
data_train.head()
```

```python
#Age数据填充中位值
data_train["Age"] = data_train["Age"].fillna(data_train["Age"].median())
data_train.head()
```

```python
#data_train.isnull().sum().sort_values(ascending=False)#查看缺失数据
#data_train.Embarked.value_counts() #查看Embarke的取值和类别
#Embarked字段填充众数
data_train["Age"] = data_train["Age"].fillna(data_train["Age"].median())
data_train['Embarked']=data_train['Embarked'].fillna('S')
data_train.isnull().sum().sort_values(ascending=False)#查看缺失数据
```

```python
#数据类型处理
#Sex字段变换数值属性,male:1,female:0
#使用map方法
sex_to_num={'male':1,'female':0}
data_train['Sex']=data_train['Sex'].map(sex_to_num)
data_train.head()
```

```python
# Embarked登船的地点也是字符串，需要变换成数字
#print (data_train["Embarked"].unique())
bark_to_num={'S':1,'C':2,'Q':3}
data_train['Embarked']=data_train['Embarked'].map(bark_to_num)
data_train.head()
```

```python
#哑变量矩阵转换，使用pandas的get_dummies函数
#性别数据dumSex
dumSex=pd.get_dummies((data_train['Sex']),prefix='Sex')
#print(dummies)
data_train=pd.concat([data_train,dumSex],axis=1)
data_train.drop('Sex',axis=1,inplace=True) #删除原Sex列
data_train.head()
```

```python
#客舱等级数据dumPclass
dumPclass=pd.get_dummies((data_train['Pclass']),prefix='Pclass')
data_train=pd.concat([data_train,dumPclass],axis=1)
data_train.drop('Pclass',axis=1,inplace=True) 
data_train.head()
```

```python
#兄弟姐妹&配偶dumSibSp dumParch
#dumSibSp=pd.get_dummies((data_train['SibSp']),prefix='SibSp')
#data_train=pd.concat([data_train,dumSibSp],axis=1)
#data_train.drop('SibSp',axis=1,inplace=True) #删除原SilbSp列

#dumParch=pd.get_dummies((data_train['Parch']),prefix='Parch')
#data_train=pd.concat([data_train,dumParch],axis=1)
#data_train.drop('Parch',axis=1,inplace=True) #删除原Parch列
#data_train.head()
```

```python
#登船地点数据dumEmbarked
dumEmbarked=pd.get_dummies((data_train['Embarked']),prefix='Embarked')
data_train=pd.concat([data_train,dumEmbarked],axis=1)
data_train.drop('Embarked',axis=1,inplace=True) #删除原Embarked列
```

```python
# 定义一个函数，从乘客姓名中提取头衔
def getTitle(name):
    str1 = name.split(',')[1]  # Mr. Owen Harris
    str2 = str1.split('.')[0]  # Mr
    return str2.strip() 

# 存放提取后的特征titleDf
titleDf = pd.DataFrame()
titleDf['Title'] = data_train['Name'].map(getTitle)
print('特征表格：',titleDf.shape)
grouped=titleDf.groupby(titleDf['Title'])
grouped.sum()
```

```python
# 姓名中头衔字符串与定义头衔类别的映射关系
title_mapDict = {
                    "Capt":       "Officer",
                    "Col":        "Officer",
                    "Major":      "Officer",
                    "Jonkheer":   "Royalty",
                    "Don":        "Royalty",
                    "Sir" :       "Royalty",
                    "Dr":         "Officer",
                    "Rev":        "Officer",
                    "the Countess":"Royalty",
                    "Dona":       "Royalty",
                    "Mme":        "Mrs",
                    "Mlle":       "Miss",
                    "Ms":         "Mrs",
                    "Mr" :        "Mr",
                    "Mrs" :       "Mrs",
                    "Miss" :      "Miss",
                    "Master" :    "Master",
                    "Lady" :      "Royalty"
                    }
# map函数：对Series每个数据应用自定义的函数计算
titleDf['Title'] = titleDf['Title'].map(title_mapDict)
titleDf = pd.get_dummies(titleDf['Title'], prefix='Title')
titleDf.head()
```

```python
# 添加one-hot编码产生的虚拟变量（dummy variables）到泰坦尼克号数据集full
data_train= pd.concat([data_train, titleDf], axis=1)
data_train.drop('Name', axis=1, inplace=True)
data_train.head()
```

```python
#后续探索：对年龄、票价进行……0-1处理
'''我们还得做一些处理，仔细看看Age和Fare两个属性，乘客的数值幅度变化，
也忒大了吧！！如果大家了解逻辑回归与梯度下降的话，会知道，
各属性值之间scale差距太大，将对收敛速度造成几万点伤害值！甚至不收敛！
所以我们先用scikit-learn里面的preprocessing模块对这俩货做一个scaling，
所谓scaling，其实就是将一些变化幅度较大的特征化到[-1,1]之内。'''
#使用相关性矩阵查看各个特征与Survived的相关系数
data_train.corr()['Survived'].sort_values(ascending=False).head(20)
# 根据各个特征与生存情况（Survived）的相关系数大小，发现下几个特征和生存情况关系最大：
# 性别（Sex）、头衔（前面所在的数据集title）、客舱等级（pclass）、船票价格（Fare）、
#家庭成员（SibSp、Parch）、登船港口（embarked）、年龄(Age)
```

```python
# 根据各个特征与生存情况（Survived）的相关系数大小，发现下几个特征和生存情况关系最大：
# 性别（Sex）、头衔（前面所在的数据集title）、客舱等级（pclass）、船票价格（Fare）、
#家庭成员（SibSp、Parch）、登船港口（embarked）、年龄(Age)
#特征选取
data_use=pd.concat([dumSex,#性别
                    titleDf,
                    dumPclass,
                    data_train['Fare'],
                    data_train['SibSp'],
                    data_train['Parch'],
                    dumEmbarked,
                    data_train['Age']#注意年龄
                   ],axis=1)
data_use.head()
```

```python
# 拆分数据集
# 训练数据集（记为source）将拆分为两部分：训练数据集（记为train：用于模型训练）和测试数据集（记为test：用于模型评估）
#拆分出来的验证数据集占20%
X = data_use[:] # 转换成矩阵形式？
Y = data_train.loc[:,'Survived']

from sklearn.cross_validation import train_test_split
train_X, test_X, train_Y, test_Y = train_test_split(X, Y, test_size=.2)
# 输出数据集大小
print ('原始数据集特征：',X.shape, 
       '训练数据集特征：',train_X.shape ,
       '验证数据集特征：',test_X.shape)

print ('原始数据集标签：',Y.shape, 
       '训练数据集标签：',train_Y.shape ,
       '验证数据集标签：',test_Y.shape)
```

```python
#开始构建模型，使用的是sklearn中的logistic回归分析模型
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
```

```python
model.fit(train_X,train_Y) #训练模型
print('模型的检验准确率为：%g' %model.score(test_X,test_Y)) #通过test_数据集检验模型的正确率
```
